```python
# STEP 1: Upload your text file
from google.colab import files
uploaded = files.upload()

# STEP 2: Import libraries
from transformers import pipeline
from math import ceil

# STEP 3: Load summarization model
summarizer = pipeline("summarization", model="t5-small")

# STEP 4: Read your uploaded file
uploaded_filename = next(iter(uploaded))  # Automatically get uploaded filename
with open("/content/" + uploaded_filename, "r", encoding="utf-8") as file:
    text = file.read().strip()

# STEP 5: Split the text into small chunks
max_chunk_size = 300  # words per chunk
words = text.split()
num_chunks = ceil(len(words) / max_chunk_size)

chunks = [' '.join(words[i * max_chunk_size:(i + 1) * max_chunk_size]) for i in range(num_chunks)]

# STEP 6: Summarize each chunk
summarized_chunks = []
for idx, chunk in enumerate(chunks):
    formatted_chunk = "summarize: " + chunk
    try:
        summary = summarizer(formatted_chunk, max_length=50, min_length=20, do_sample=False)
        summarized_chunks.append(summary[0]['summary_text'])
        print(f"✅ Chunk {idx+1}/{num_chunks} summarized.")
    except Exception as e:
        print(f"❌ Error summarizing chunk {idx+1}: {e}")
        summarized_chunks.append("[Summary failed for this chunk]")

# STEP 7: Combine all summaries
final_summary = '\n'.join(summarized_chunks)

# STEP 8: Save the final summary
with open("summary.txt", "w", encoding="utf-8") as file:
    file.write(final_summary)

print("\n🎯 Final summary created successfully!")

# STEP 9: Download the file
files.download('summary.txt')
```
